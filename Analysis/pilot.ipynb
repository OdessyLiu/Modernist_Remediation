# Clean and merge collected review files

```python
# Read and clean scraped review data (Raw)
# Review(1)

reviewPath = "../Data/Web-Scraped_Data/Raw/AllReviews/MainDataset/CSV/AllReviews(1).csv"
review = pd.read_csv(reviewPath)
```

```python
review.head()
```

```python
review.describe()
```

```python
# review, old data

reviewOldPath = "../Data/Web-Scraped_Data/Raw/AllReviews/MainDataset/CSV/AllReviews.csv"
reviewOld = pd.read_csv(reviewOldPath)
reviewOld.describe()
```

```python
# Unique titles in review

uniqueTitlesReview = review['Title'].unique().tolist()
uniqueTitlesReview = [title.replace('\ufeff', '') for title in uniqueTitlesReview]

uniqueTitlesReview = list(set(uniqueTitlesReview))

print(uniqueTitlesReview)
print(len(uniqueTitlesReview))
```

```python
# Unique titles in reviewOld

uniqueTitlesReviewOld = reviewOld['Title'].unique().tolist()
uniqueTitlesReviewOld = [title.replace('\ufeff', '') for title in uniqueTitlesReviewOld]

uniqueTitlesReviewOld = list(set(uniqueTitlesReviewOld))

print(uniqueTitlesReviewOld)
print(len(uniqueTitlesReviewOld))
```

```python
# if there are any common titles between review and reviewOld

commonTitles = list(set(uniqueTitlesReview) & set(uniqueTitlesReviewOld))
commonTitles
```

Thus, the AllReviews dataset (AllReviews and AllReviews(1)) contains the reviews of 135 graphic comics.

```python
# Merge the two reveiw data

reviewMerged = pd.concat([review, reviewOld]).drop_duplicates().reset_index(drop=True)
reviewMerged["Title"] = reviewMerged["Title"].str.replace('\ufeff', '')
reviewMerged.describe()
```

```python
# drop duplicates

reviewMerged.drop_duplicates()
reviewMerged.describe()
```

```python
duplicated_indices = reviewMerged[reviewMerged.duplicated(subset=['Title', 'Review_Author', 'Review_Text'])].index
duplicated_indices
```

```python
duplicated_df = reviewMerged.loc[duplicated_indices]
duplicated_df
```

```python
reviewMerged = reviewMerged.drop(duplicated_indices)
reviewMerged.describe()
```

```python
# save reviewMerged into a CSV file

reviewMergedPath = "../Data/Web-Scraped_Data/AllReviews_135_Unique.csv"
reviewMerged.to_csv(reviewMergedPath, index=False)
```

```python
titlesCollected = reviewMerged['Title'].unique().tolist()
titlesCollected = [title.lower() for title in titlesCollected]

titlesInBio = bioCleaned['Title'].tolist()
titlesInBio = [title.lower() for title in titlesInBio]

titlesNotInBio = [title for title in titlesCollected if title not in titlesInBio]
len(titlesNotInBio)
```
